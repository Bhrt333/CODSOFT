## Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
plt.scatter(ad['TV'],ad['Sales'] ,color= 'green')
plt.title('Sales based on TV ADs')
plt.xlabel('TV Advertisment')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(ad['Radio'],ad['Sales'] ,color= 'brown')
plt.title('Sales based on Radio ADs')
plt.xlabel('Radio Advertisment')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(ad['Newspaper'],ad['Sales'] ,color= 'blue')
plt.title('Sales based on Newspaper ADs')
plt.xlabel('NewspaperAdvertisment')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

**In this plots we can find only TV and Radio ads showing some patterns.**

## Model Selection

```python
from sklearn.linear_model import LinearRegression
```

```python
x = ad[['TV','Radio','Newspaper']] # indipendent
y = ad['Sales'] #dependent
```

```python
model = LinearRegression()
model = model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
print('Prediction ', y_pred)
```

```python
scores = model.score(x,y)
print(scores)
```
